```python
import os
import shutil
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import kagglehub
import zipfile
import glob

# Define paths
dataset_dir = r"C:\Users\jidub\Downloads\Dataset_Extracted\My%20Dataset"
print(f"Using main dataset directory: {dataset_dir}")

# Download Kaggle datasets
print("\nDownloading additional datasets from Kaggle...")
try:
    kaggle_dataset1 = kagglehub.dataset_download("andrewmvd/pothole-detection")
    print(f"Downloaded first Kaggle dataset to: {kaggle_dataset1}")
    
    kaggle_dataset2 = kagglehub.dataset_download("sachinpatel21/pothole-image-dataset")
    print(f"Downloaded second Kaggle dataset to: {kaggle_dataset2}")
except Exception as e:
    print(f"Error downloading Kaggle datasets: {e}")
    print("Continuing with local dataset only...")
    kaggle_dataset1 = None
    kaggle_dataset2 = None

# Create output directories for our processed dataset
output_dir = os.path.join(os.path.dirname(dataset_dir), 'processed_dataset')
processed_train_dir = os.path.join(output_dir, 'train')
processed_test_dir = os.path.join(output_dir, 'test')

# Create directories if they don't exist
for class_name in ['pothole', 'non-pothole']:
    os.makedirs(os.path.join(processed_train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(processed_test_dir, class_name), exist_ok=True)

# Process main dataset
print("\nProcessing main dataset...")
train_dir = os.path.join(dataset_dir, 'train')
test_dir = os.path.join(dataset_dir, 'test')

# Map the existing folders to what our model expects
class_mapping = {
    'Plain': 'non-pothole',
    'Pothole': 'pothole'
}

# Function to copy images
def copy_images(src_dir, dest_dir, original_class, new_class):
    original_class_dir = os.path.join(src_dir, original_class)
    if os.path.exists(original_class_dir):
        new_class_dir = os.path.join(dest_dir, new_class)
        img_count = 0
        for img_name in os.listdir(original_class_dir):
            if img_name.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                src_path = os.path.join(original_class_dir, img_name)
                # Add a prefix to avoid filename collisions
                dst_path = os.path.join(new_class_dir, f"main_{original_class}_{img_name}")
                shutil.copy(src_path, dst_path)
                img_count += 1
        print(f"Copied {img_count} images from {src_dir}/{original_class} to {dest_dir}/{new_class}")

# Copy training images
if os.path.exists(train_dir):
    for original_class, new_class in class_mapping.items():
        copy_images(train_dir, processed_train_dir, original_class, new_class)

# Copy testing images
if os.path.exists(test_dir):
    for original_class, new_class in class_mapping.items():
        copy_images(test_dir, processed_test_dir, original_class, new_class)

# Function to process Kaggle dataset 1 (andrewmvd/pothole-detection)
def process_kaggle_dataset1(dataset_path):
    print("\nProcessing first Kaggle dataset...")
    
    # Find all image files recursively
    image_files = []
    for ext in ['*.jpg', '*.jpeg', '*.png', '*.bmp']:
        image_files.extend(glob.glob(os.path.join(dataset_path, '**', ext), recursive=True))
    
    print(f"Found {len(image_files)} images in first Kaggle dataset")
    
    if not image_files:
        return 0, 0
    
    # This dataset has annotations in XML files, so we need to check folders and filenames
    pothole_count = 0
    non_pothole_count = 0
    
    # For simplicity, we'll check paths and filenames for indicators
    # In a real-world scenario, you'd parse the XML files for proper annotations
    for img_path in image_files:
        img_name = os.path.basename(img_path).lower()
        parent_dir = os.path.basename(os.path.dirname(img_path)).lower()
        
        # Check if path contains pothole indicators
        if 'pothole' in img_name or 'pothole' in parent_dir:
            # 80% to training, 20% to testing (simple split)
            if np.random.rand() < 0.8:
                dest_dir = os.path.join(processed_train_dir, 'pothole')
                prefix = 'k1_train_pothole_'
            else:
                dest_dir = os.path.join(processed_test_dir, 'pothole')
                prefix = 'k1_test_pothole_'
            pothole_count += 1
        else:
            # Assume non-pothole if not explicitly labeled
            if np.random.rand() < 0.8:
                dest_dir = os.path.join(processed_train_dir, 'non-pothole')
                prefix = 'k1_train_nonpothole_'
            else:
                dest_dir = os.path.join(processed_test_dir, 'non-pothole')
                prefix = 'k1_test_nonpothole_'
            non_pothole_count += 1
            
        # Copy the image file
        dest_path = os.path.join(dest_dir, f"{prefix}{img_name}")
        try:
            shutil.copy(img_path, dest_path)
        except Exception as e:
            print(f"Error copying {img_path}: {e}")
    
    print(f"Processed {pothole_count} pothole and {non_pothole_count} non-pothole images from first Kaggle dataset")
    return pothole_count, non_pothole_count

# Function to process Kaggle dataset 2 (sachinpatel21/pothole-image-dataset)
def process_kaggle_dataset2(dataset_path):
    print("\nProcessing second Kaggle dataset...")
    
    # Find all image files recursively
    image_files = []
    for ext in ['*.jpg', '*.jpeg', '*.png', '*.bmp']:
        image_files.extend(glob.glob(os.path.join(dataset_path, '**', ext), recursive=True))
    
    print(f"Found {len(image_files)} images in second Kaggle dataset")
    
    if not image_files:
        return 0, 0
    
    pothole_count = 0
    non_pothole_count = 0
    
    # This dataset likely has a specific structure - let's detect it
    for img_path in image_files:
        img_name = os.path.basename(img_path).lower()
        parent_dir = os.path.basename(os.path.dirname(img_path)).lower()
        
        # Check if path contains pothole indicators
        if 'pothole' in img_name or 'pothole' in parent_dir:
            # 80% to training, 20% to testing
            if np.random.rand() < 0.8:
                dest_dir = os.path.join(processed_train_dir, 'pothole')
                prefix = 'k2_train_pothole_'
            else:
                dest_dir = os.path.join(processed_test_dir, 'pothole')
                prefix = 'k2_test_pothole_'
            pothole_count += 1
        else:
            # Assume non-pothole if not explicitly labeled
            if np.random.rand() < 0.8:
                dest_dir = os.path.join(processed_train_dir, 'non-pothole')
                prefix = 'k2_train_nonpothole_'
            else:
                dest_dir = os.path.join(processed_test_dir, 'non-pothole')
                prefix = 'k2_test_nonpothole_'
            non_pothole_count += 1
            
        # Copy the image file
        dest_path = os.path.join(dest_dir, f"{prefix}{img_name}")
        try:
            shutil.copy(img_path, dest_path)
        except Exception as e:
            print(f"Error copying {img_path}: {e}")
    
    print(f"Processed {pothole_count} pothole and {non_pothole_count} non-pothole images from second Kaggle dataset")
    return pothole_count, non_pothole_count

# Process Kaggle datasets if they were downloaded successfully
kaggle1_pothole = 0
kaggle1_non_pothole = 0
kaggle2_pothole = 0
kaggle2_non_pothole = 0

if kaggle_dataset1:
    kaggle1_pothole, kaggle1_non_pothole = process_kaggle_dataset1(kaggle_dataset1)

if kaggle_dataset2:
    kaggle2_pothole, kaggle2_non_pothole = process_kaggle_dataset2(kaggle_dataset2)

# Count images in our processed directories
train_pothole_count = len(os.listdir(os.path.join(processed_train_dir, 'pothole'))) if os.path.exists(os.path.join(processed_train_dir, 'pothole')) else 0
train_non_pothole_count = len(os.listdir(os.path.join(processed_train_dir, 'non-pothole'))) if os.path.exists(os.path.join(processed_train_dir, 'non-pothole')) else 0
test_pothole_count = len(os.listdir(os.path.join(processed_test_dir, 'pothole'))) if os.path.exists(os.path.join(processed_test_dir, 'pothole')) else 0
test_non_pothole_count = len(os.listdir(os.path.join(processed_test_dir, 'non-pothole'))) if os.path.exists(os.path.join(processed_test_dir, 'non-pothole')) else 0

print("\nFinal dataset statistics:")
print(f"Training: {train_pothole_count} pothole, {train_non_pothole_count} non-pothole")
print(f"Testing: {test_pothole_count} pothole, {test_non_pothole_count} non-pothole")
print(f"Total images: {train_pothole_count + train_non_pothole_count + test_pothole_count + test_non_pothole_count}")

# Check if we have enough data to proceed
if (train_pothole_count == 0 or train_non_pothole_count == 0 or 
    test_pothole_count == 0 or test_non_pothole_count == 0):
    print("\nERROR: Not enough images found in categories. Please check your dataset structure.")
    import sys
    sys.exit(1)

# Define image dimensions
img_width, img_height = 224, 224
batch_size = 32  # Increased batch size as we have more data

# Calculate class weights to help with imbalance
total_train = train_pothole_count + train_non_pothole_count
weight_for_0 = (1 / train_non_pothole_count) * (total_train / 2.0) if train_non_pothole_count > 0 else 1.0
weight_for_1 = (1 / train_pothole_count) * (total_train / 2.0) if train_pothole_count > 0 else 1.0
class_weight = {0: weight_for_0, 1: weight_for_1}
print(f"\nUsing class weights: {class_weight}")

# Data augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest',
    validation_split=0.1  # Use 10% of training data as validation set
)

# Only rescaling for testing
test_datagen = ImageDataGenerator(rescale=1./255)

print("\nLoading training and validation data...")
# Load training data
train_generator = train_datagen.flow_from_directory(
    processed_train_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)

# Load validation data (split from training)
validation_generator = train_datagen.flow_from_directory(
    processed_train_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)

print("\nLoading testing data...")
# Load testing data
test_generator = test_datagen.flow_from_directory(
    processed_test_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

# Build the model
def build_model():
    # Use a more modern architecture with increased regularization
    base_model = tf.keras.applications.MobileNetV2(
        input_shape=(img_width, img_height, 3),
        include_top=False,
        weights='imagenet'
    )
    
    # Freeze the base model
    base_model.trainable = False
    
    # Add custom layers
    inputs = tf.keras.Input(shape=(img_width, img_height, 3))
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    
    model = models.Model(inputs=inputs, outputs=outputs)
    return model

# Create model
model = build_model()

# Compile model
model.compile(
    loss='binary_crossentropy',
    optimizer=optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

# Print model summary
model.summary()

# Set up callbacks
checkpoint_path = 'best_pothole_model.h5'
checkpoint = ModelCheckpoint(
    checkpoint_path,
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=0.00001,
    verbose=1
)

# Train the model
epochs = 20

# Calculate steps
steps_per_epoch = max(1, train_generator.samples // batch_size)
validation_steps = max(1, validation_generator.samples // batch_size)

print(f"\nTraining model with {train_generator.samples} training samples and {validation_generator.samples} validation samples")
print(f"Steps per epoch: {steps_per_epoch}, Validation steps: {validation_steps}")

# First phase: train only the top layers
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    class_weight=class_weight,
    callbacks=[checkpoint, early_stopping, reduce_lr]
)

# Second phase: fine-tune the last few layers of the base model
print("\nFine-tuning the model...")
# Unfreeze the last 30 layers
base_model = model.layers[1]
base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False

# Recompile with lower learning rate
model.compile(
    loss='binary_crossentropy',
    optimizer=optimizers.Adam(learning_rate=0.0001),
    metrics=['accuracy']
)

# Continue training
fine_tune_epochs = 10
total_epochs = epochs + fine_tune_epochs

history_fine = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=total_epochs,
    initial_epoch=history.epoch[-1] + 1,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    class_weight=class_weight,
    callbacks=[checkpoint, early_stopping, reduce_lr]
)

# Combine histories
history_combined = {}
for key in history.history.keys():
    history_combined[key] = history.history[key] + history_fine.history.get(key, [])

# Plot training history
def plot_training_history(history):
    plt.figure(figsize=(12, 5))
    
    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.get('accuracy', []))
    plt.plot(history.get('val_accuracy', []))
    plt.axvline(x=epochs, color='r', linestyle='--')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation', 'Start Fine Tuning'], loc='lower right')
    
    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.get('loss', []))
    plt.plot(history.get('val_loss', []))
    plt.axvline(x=epochs, color='r', linestyle='--')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation', 'Start Fine Tuning'], loc='upper right')
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

plot_training_history(history_combined)

# Load the best model
best_model = tf.keras.models.load_model(checkpoint_path)

# Evaluate the model
print("\nEvaluating model...")
evaluation = best_model.evaluate(test_generator)
print(f"Test Loss: {evaluation[0]:.4f}")
print(f"Test Accuracy: {evaluation[1]:.4f}")

# Generate predictions
test_generator.reset()
y_pred = best_model.predict(test_generator, steps=int(np.ceil(test_generator.samples / batch_size)))
y_pred_binary = (y_pred > 0.5).astype(int)

# Get true labels
y_true = test_generator.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred_binary)
plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=test_generator.class_indices.keys(),
            yticklabels=test_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()

# Classification Report
print("\nClassification Report:")
print(classification_report(y_true, y_pred_binary, 
                          target_names=list(test_generator.class_indices.keys())))

# Save the model
best_model.save('pothole_detection_model_final.h5')
print("Final model saved as 'pothole_detection_model_final.h5'")

# Function to make predictions on new images
def predict_image(image_path, model):
    from tensorflow.keras.preprocessing import image
    
    img = image.load_img(image_path, target_size=(img_width, img_height))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    prediction = model.predict(img_array)[0][0]
    result = "Pothole" if prediction > 0.5 else "Non-Pothole"
    confidence = prediction if prediction > 0.5 else 1 - prediction
    
    return result, confidence

# Create a function to test with some sample images
def test_with_samples(sample_dir, model, num_samples=5):
    import random
    
    all_images = []
    for class_name in os.listdir(sample_dir):
        class_dir = os.path.join(sample_dir, class_name)
        if os.path.isdir(class_dir):
            for img_name in os.listdir(class_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    img_path = os.path.join(class_dir, img_name)
                    all_images.append((img_path, class_name))
    
    # Select random samples
    if len(all_images) > num_samples:
        samples = random.sample(all_images, num_samples)
    else:
        samples = all_images
    
    # Display results
    plt.figure(figsize=(15, 10))
    for i, (img_path, true_class) in enumerate(samples):
        # Predict
        result, confidence = predict_image(img_path, best_model)
        
        # Display
        img = plt.imread(img_path)
        plt.subplot(1, len(samples), i+1)
        plt.imshow(img)
        plt.title(f"True: {true_class}\nPred: {result}\nConf: {confidence:.2%}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.savefig('sample_predictions.png')
    plt.show()

# Test with sample images from the test directory
print("\nTesting with sample images...")
test_with_samples(processed_test_dir, best_model)

# Calculate class activation maps for sample images
def generate_grad_cam(img_path, model, last_conv_layer_name='Conv_1'):
    from tensorflow.keras.preprocessing import image
    import cv2
    import tensorflow as tf
    
    # Load and preprocess image
    img = image.load_img(img_path, target_size=(img_width, img_height))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    # Create Grad-CAM model
    grad_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )
    
    # Compute gradients
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        class_idx = 0 if predictions[0][0] < 0.5 else 1
        loss = predictions[:, 0]
    
    # Extract feature outputs
    output = conv_outputs[0]
    grads = tape.gradient(loss, conv_outputs)[0]
    
    # Pool gradients
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Weight output feature maps with gradients
    heatmap = output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    # Normalize heatmap
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    heatmap = heatmap.numpy()
    
    # Resize and convert to RGB
    heatmap = cv2.resize(heatmap, (img_width, img_height))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    
    # Convert image to BGR (for OpenCV)
    img_bgr = cv2.cvtColor(img_array[0] * 255, cv2.COLOR_RGB2BGR)
    
    # Superimpose heatmap
    superimposed_img = cv2.addWeighted(img_bgr, 0.6, heatmap, 0.4, 0)
    
    return superimposed_img

# Display Grad-CAM for sample images
def show_grad_cam_samples(sample_dir, model, num_samples=3):
    import random
    import cv2
    
    # Find last convolutional layer
    last_conv_layer = None
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            last_conv_layer = layer.name
            break
    
    if not last_conv_layer:
        print("Could not find convolutional layer for Grad-CAM")
        return
        
    print(f"Using layer {last_conv_layer} for Grad-CAM")
    
    # Get sample images
    all_images = []
    for class_name in os.listdir(sample_dir):
        class_dir = os.path.join(sample_dir, class_name)
        if os.path.isdir(class_dir):
            img_paths = [os.path.join(class_dir, f) for f in os.listdir(class_dir) 
                       if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
            if img_paths:
                # Get up to 2 images per class
                samples = random.sample(img_paths, min(2, len(img_paths)))
                for img_path in samples:
                    all_images.append((img_path, class_name))
    
    if not all_images:
        print("No images found for Grad-CAM")
        return
        
    # Select random samples if we have too many
    if len(all_images) > num_samples:
        samples = random.sample(all_images, num_samples)
    else:
        samples = all_images
    
    # Generate and display Grad-CAM
    plt.figure(figsize=(15, 5 * len(samples)))
    for i, (img_path, true_class) in enumerate(samples):
        # Get prediction
        result, confidence = predict_image(img_path, model)
        
        # Generate Grad-CAM
        try:
            cam_image = generate_grad_cam(img_path, model, last_conv_layer)
            
            # Original image
            plt.subplot(len(samples), 2, 2*i+1)
            orig_img = plt.imread(img_path)
            plt.imshow(orig_img)
            plt.title(f"Original - Class: {true_class}")
            plt.axis('off')
            
            # Grad-CAM image
            plt.subplot(len(samples), 2, 2*i+2)
            plt.imshow(cv2.cvtColor(cam_image, cv2.COLOR_BGR2RGB))
            plt.title(f"Grad-CAM - Pred: {result} ({confidence:.2%})")
            plt.axis('off')
        except Exception as e:
            print(f"Error generating Grad-CAM for {img_path}: {e}")
    
    plt.tight_layout()
    plt.savefig('grad_cam_visualization.png')
    plt.show()

print("\nGenerating Grad-CAM visualizations...")
try:
    show_grad_cam_samples(processed_test_dir, best_model)
except Exception as e:
    print(f"Could not generate Grad-CAM visualizations: {e}")

print("\nPothole detection system training and evaluation complete.")
```

```python
import os
import numpy as np
import tensorflow as tf
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import time
import glob
from skimage.segmentation import flood_fill, flood

# Constants
IMG_WIDTH, IMG_HEIGHT = 224, 224  # Same dimensions used during training
MODEL_PATH = 'pothole_detection_model_final.h5'  # Path to your trained model

def load_pothole_model(model_path):
    """Load the trained pothole detection model"""
    print(f"Loading model from {model_path}...")
    
    try:
        model = load_model(model_path)
        print("Model loaded successfully!")
        return model
    except Exception as e:
        print(f"Error loading model: {e}")
        return None

def preprocess_image(img):
    """Preprocess an image for prediction"""
    # Resize image
    img_resized = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))
    
    # Normalize pixel values
    img_normalized = img_resized / 255.0
    
    # Expand dimensions to match model input shape
    img_batch = np.expand_dims(img_normalized, axis=0)
    
    return img_batch

def detect_potholes(frame, model, confidence_threshold=0.5):
    """
    Detect potholes in an image using a sliding window approach
    Returns list of (confidence, x, y, w, h) for each detected pothole
    """
    original_height, original_width = frame.shape[:2]
    detections = []
    
    # First, check if the entire frame has a pothole
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame_preprocessed = preprocess_image(frame_rgb)
    whole_frame_prediction = model.predict(frame_preprocessed, verbose=0)[0][0]
    
    # If no pothole in the whole frame, return empty list
    if whole_frame_prediction < confidence_threshold:
        return detections
    
    # Otherwise, use sliding window to locate pothole regions
    # Define window sizes as fractions of the image
    window_sizes = [
        (original_width // 2, original_height // 2),
        (original_width // 3, original_height // 3),
        (original_width // 4, original_height // 4)
    ]
    
    # Define strides as fraction of window size
    stride_ratio = 0.5
    
    for window_width, window_height in window_sizes:
        stride_x = int(window_width * stride_ratio)
        stride_y = int(window_height * stride_ratio)
        
        for y in range(0, original_height - window_height + 1, stride_y):
            for x in range(0, original_width - window_width + 1, stride_x):
                # Extract window
                window = frame[y:y+window_height, x:x+window_width]
                
                # Skip if window is too small
                if window.shape[0] < 10 or window.shape[1] < 10:
                    continue
                
                # Preprocess and predict
                window_rgb = cv2.cvtColor(window, cv2.COLOR_BGR2RGB)
                window_preprocessed = preprocess_image(window_rgb)
                prediction = model.predict(window_preprocessed, verbose=0)[0][0]
                
                # If pothole detected with high confidence
                if prediction > confidence_threshold:
                    detections.append((prediction, x, y, window_width, window_height))
    
    # Non-maximum suppression to remove overlapping detections
    detections = non_max_suppression(detections, 0.3)  # 0.3 is the IoU threshold
    
    return detections

def non_max_suppression(detections, iou_threshold):
    """
    Apply non-maximum suppression to avoid duplicate detections
    """
    if len(detections) == 0:
        return []
    
    # Sort by confidence
    detections.sort(key=lambda x: x[0], reverse=True)
    
    # Initialize list of selected detections
    selected = []
    
    # Do NMS
    while len(detections) > 0:
        # Take the detection with highest confidence
        current = detections.pop(0)
        selected.append(current)
        
        # Remove overlapping detections
        non_overlapping = []
        for detection in detections:
            # Calculate IoU between current and detection
            _, x1, y1, w1, h1 = current
            _, x2, y2, w2, h2 = detection
            
            # Calculate intersection
            x_left = max(x1, x2)
            y_top = max(y1, y2)
            x_right = min(x1 + w1, x2 + w2)
            y_bottom = min(y1 + h1, y2 + h2)
            
            if x_right < x_left or y_bottom < y_top:
                intersection_area = 0
            else:
                intersection_area = (x_right - x_left) * (y_bottom - y_top)
            
            # Calculate areas
            area1 = w1 * h1
            area2 = w2 * h2
            
            # Calculate IoU
            iou = intersection_area / float(area1 + area2 - intersection_area)
            
            # If IoU is less than threshold, keep the detection
            if iou < iou_threshold:
                non_overlapping.append(detection)
        
        detections = non_overlapping
    
    return selected

def generate_heatmap(frame, model):
    """
    Generate a heatmap showing where potholes are likely to be
    """
    original_height, original_width = frame.shape[:2]
    heatmap = np.zeros((original_height, original_width), dtype=np.float32)
    
    # Use smaller windows for finer detection, but more processing time
    window_sizes = [
        (original_width // 4, original_height // 4),
        (original_width // 6, original_height // 6)
    ]
    
    # Define strides as fraction of window size
    stride_ratio = 0.5
    
    for window_width, window_height in window_sizes:
        stride_x = int(window_width * stride_ratio)
        stride_y = int(window_height * stride_ratio)
        
        for y in range(0, original_height - window_height + 1, stride_y):
            for x in range(0, original_width - window_width + 1, stride_x):
                # Extract window
                window = frame[y:y+window_height, x:x+window_width]
                
                # Skip if window is too small
                if window.shape[0] < 10 or window.shape[1] < 10:
                    continue
                
                # Preprocess and predict
                window_rgb = cv2.cvtColor(window, cv2.COLOR_BGR2RGB)
                window_preprocessed = preprocess_image(window_rgb)
                prediction = model.predict(window_preprocessed, verbose=0)[0][0]
                
                # Add prediction to heatmap
                heatmap[y:y+window_height, x:x+window_width] += prediction
    
    # Normalize heatmap to 0-1 range
    if np.max(heatmap) > 0:
        heatmap = heatmap / np.max(heatmap)
    
    return heatmap

def predict_image_file(image_path, model):
    """Predict a pothole in a single image file with bounding box"""
    # Load image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not load image {image_path}")
        return None, 0
    
    # Convert from BGR to RGB (CV2 loads as BGR, but our model expects RGB)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # First, check if the whole image has a pothole
    start_time = time.time()
    img_preprocessed = preprocess_image(img_rgb)
    prediction = model.predict(img_preprocessed)[0][0]
    end_time = time.time()
    
    # Process result
    is_pothole = prediction > 0.5
    confidence = prediction if is_pothole else 1 - prediction
    result_text = "Pothole" if is_pothole else "Non-Pothole"
    
    # Generate a copy for drawing
    img_with_detection = img.copy()
    
    # If it's a pothole, try to locate it more precisely
    if is_pothole:
        # Detect potholes with bounding boxes
        detections = detect_potholes(img, model, confidence_threshold=0.6)
        
        if detections:
            # Draw bounding boxes for each detection
            for detection in detections:
                conf, x, y, w, h = detection
                # Draw rectangle
                cv2.rectangle(img_with_detection, (x, y), (x+w, y+h), (0, 0, 255), 2)
                # Add confidence text
                cv2.putText(img_with_detection, f"{conf:.2%}", (x, y-10),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        else:
            # If no specific regions detected, draw full image box
            h, w = img.shape[:2]
            cv2.rectangle(img_with_detection, (0, 0), (w, h), (0, 0, 255), 2)
            
        # Generate heatmap for visualization
        heatmap = generate_heatmap(img, model)
        heatmap_colored = cv2.applyColorMap((heatmap * 255).astype(np.uint8), cv2.COLORMAP_JET)
        
        # Overlay heatmap on original image
        overlay = cv2.addWeighted(img, 0.7, heatmap_colored, 0.3, 0)
    
    # Add text with overall prediction
    cv2.putText(img_with_detection, f"{result_text}: {confidence:.2%}", (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255) if is_pothole else (0, 255, 0), 2)
    cv2.putText(img_with_detection, f"Inference time: {(end_time - start_time) * 1000:.2f} ms", 
                (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
    
    print(f"Prediction: {result_text} with {confidence:.2%} confidence")
    print(f"Inference time: {(end_time - start_time) * 1000:.2f} ms")
    
    # Display results
    plt.figure(figsize=(12, 6))
    
    # Original with boxes
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_with_detection, cv2.COLOR_BGR2RGB))
    plt.title(f"Detection: {result_text} ({confidence:.2%})")
    plt.axis('off')
    
    # Heatmap overlay if it's a pothole
    plt.subplot(1, 2, 2)
    if is_pothole:
        plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
        plt.title("Pothole Heatmap")
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title("Original Image")
    plt.axis('off')
    
    # Save results
    save_path = os.path.join('results', os.path.basename(image_path))
    os.makedirs('results', exist_ok=True)
    plt.savefig(save_path)
    plt.show()
    
    # Also save a version with OpenCV
    cv2.imwrite(os.path.join('results', f"cv2_{os.path.basename(image_path)}"), img_with_detection)
    if is_pothole:
        cv2.imwrite(os.path.join('results', f"heatmap_{os.path.basename(image_path)}"), overlay)
    
    return is_pothole, confidence

def process_video(video_path, model, output_path=None, confidence_threshold=0.6, process_every_n_frames=2):
    """Process a video file for pothole detection with bounding boxes"""
    # Open video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video {video_path}")
        return
    
    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Create output video writer
    if output_path is None:
        output_path = os.path.join('results', f"output_{os.path.basename(video_path)}")
    
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # or 'XVID' for AVI
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))
    
    frame_count = 0
    pothole_frames = 0
    processing_times = []
    
    # For tracking potholes across frames (simple smoothing)
    prev_detections = []
    detection_smoothing_alpha = 0.7  # Higher means more weight on new detections
    
    print(f"Processing video: {video_path}")
    print(f"Total frames: {total_frames}, FPS: {fps:.2f}")
    print("Press Ctrl+C to stop if needed.")
    
    start_time_total = time.time()
    
    try:
        # Process the video
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            # Process every N frames to improve speed
            process_this_frame = (frame_count % process_every_n_frames == 0)
            frame_with_boxes = frame.copy()
            
            if process_this_frame:
                # Detect potholes
                start_time = time.time()
                
                # First check if the entire frame has a pothole
                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                frame_preprocessed = preprocess_image(frame_rgb)
                prediction = model.predict(frame_preprocessed, verbose=0)[0][0]
                is_pothole = prediction > confidence_threshold
                
                if is_pothole:
                    # Detect pothole locations
                    detections = detect_potholes(frame, model, confidence_threshold)
                    
                    # Simple smoothing across frames
                    if prev_detections and detections:
                        # Weight previous detections to reduce flickering
                        smoothed_detections = []
                        for detection in detections:
                            # Find closest previous detection
                            best_match = None
                            best_iou = 0
                            for prev_detection in prev_detections:
                                # Calculate IoU
                                _, x1, y1, w1, h1 = detection
                                _, x2, y2, w2, h2 = prev_detection
                                
                                # Calculate intersection
                                x_left = max(x1, x2)
                                y_top = max(y1, y2)
                                x_right = min(x1 + w1, x2 + w2)
                                y_bottom = min(y1 + h1, y2 + h2)
                                
                                if x_right < x_left or y_bottom < y_top:
                                    intersection_area = 0
                                else:
                                    intersection_area = (x_right - x_left) * (y_bottom - y_top)
                                
                                # Calculate areas
                                area1 = w1 * h1
                                area2 = w2 * h2
                                
                                # Calculate IoU
                                iou = intersection_area / float(area1 + area2 - intersection_area)
                                
                                if iou > best_iou:
                                    best_iou = iou
                                    best_match = prev_detection
                            
                            if best_match and best_iou > 0.3:
                                # Smooth the detection
                                conf1, x1, y1, w1, h1 = detection
                                conf2, x2, y2, w2, h2 = best_match
                                
                                # Weighted average
                                new_conf = detection_smoothing_alpha * conf1 + (1 - detection_smoothing_alpha) * conf2
                                new_x = int(detection_smoothing_alpha * x1 + (1 - detection_smoothing_alpha) * x2)
                                new_y = int(detection_smoothing_alpha * y1 + (1 - detection_smoothing_alpha) * y2)
                                new_w = int(detection_smoothing_alpha * w1 + (1 - detection_smoothing_alpha) * w2)
                                new_h = int(detection_smoothing_alpha * h1 + (1 - detection_smoothing_alpha) * h2)
                                
                                smoothed_detections.append((new_conf, new_x, new_y, new_w, new_h))
                            else:
                                smoothed_detections.append(detection)
                        
                        detections = smoothed_detections
                    
                    # Update previous detections
                    prev_detections = detections.copy()
                    
                    # Draw bounding boxes
                    for detection in detections:
                        conf, x, y, w, h = detection
                        # Draw rectangle
                        cv2.rectangle(frame_with_boxes, (x, y), (x+w, y+h), (0, 0, 255), 2)
                        # Add confidence text
                        cv2.putText(frame_with_boxes, f"Pothole: {conf:.2%}", 
                                  (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
                    
                    pothole_frames += 1
                
                end_time = time.time()
                processing_time = end_time - start_time
                processing_times.append(processing_time)
                
                # Add general status info
                cv2.putText(frame_with_boxes, 
                          f"Status: {'POTHOLE DETECTED' if is_pothole else 'NO POTHOLE'}", 
                          (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, 
                          (0, 0, 255) if is_pothole else (0, 255, 0), 2)
                          
                cv2.putText(frame_with_boxes, 
                          f"Confidence: {prediction:.2%}", 
                          (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, 
                          (0, 0, 255) if is_pothole else (0, 255, 0), 2)
                          
                cv2.putText(frame_with_boxes, 
                          f"Frame: {frame_count}/{total_frames}", 
                          (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
            else:
                # For frames we skip processing, use the previous frame's detections
                if prev_detections:
                    for detection in prev_detections:
                        conf, x, y, w, h = detection
                        # Draw rectangle
                        cv2.rectangle(frame_with_boxes, (x, y), (x+w, y+h), (0, 0, 255), 2)
                        # Add confidence text
                        cv2.putText(frame_with_boxes, f"Pothole: {conf:.2%}", 
                                  (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
                
                # Add frame counter
                cv2.putText(frame_with_boxes, 
                          f"Frame: {frame_count}/{total_frames}", 
                          (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
            
            # Write frame to output video
            out.write(frame_with_boxes)
            
            # Print progress periodically
            if frame_count % 100 == 0:
                elapsed = time.time() - start_time_total
                estimated_total = (elapsed / (frame_count + 1)) * total_frames if frame_count > 0 else 0
                remaining = max(0, estimated_total - elapsed)
                print(f"Processed {frame_count}/{total_frames} frames ({frame_count/total_frames*100:.1f}%) - ETA: {remaining/60:.1f} min")
            
            frame_count += 1
            
    except KeyboardInterrupt:
        print("Processing interrupted by user.")
        
    except Exception as e:
        print(f"Error during video processing: {e}")
        
    finally:
        # Release resources
        cap.release()
        out.release()
        
        # Print statistics
        elapsed_total = time.time() - start_time_total
        avg_processing_time = sum(processing_times) / len(processing_times) if processing_times else 0
        print(f"\nVideo processing complete in {elapsed_total/60:.2f} minutes")
        print(f"  Total frames processed: {frame_count}")
        print(f"  Frames with potholes: {pothole_frames} ({pothole_frames/frame_count:.2%})")
        print(f"  Average processing time: {avg_processing_time*1000:.2f} ms per frame")
        print(f"  Effective FPS: {1/avg_processing_time:.2f}")
        
        print(f"Output saved to: {output_path}")
        
        # Save a summary report
        report_path = os.path.join('results', f"report_{os.path.basename(video_path)}.txt")
        with open(report_path, 'w') as f:
            f.write(f"Video Processing Report for {video_path}\n")
            f.write(f"Total processing time: {elapsed_total/60:.2f} minutes\n")
            f.write(f"Total frames: {frame_count}\n")
            f.write(f"Frames with potholes: {pothole_frames} ({pothole_frames/frame_count:.2%})\n")
            f.write(f"Average processing time: {avg_processing_time*1000:.2f} ms per frame\n")
            f.write(f"Effective FPS: {1/avg_processing_time:.2f}\n")
        
        print(f"Report saved to: {report_path}")

def process_image_folder(folder_path, model):
    """Process all images in a folder"""
    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
    images = []
    
    # Find all images in the folder
    for file in os.listdir(folder_path):
        if any(file.lower().endswith(ext) for ext in image_extensions):
            images.append(os.path.join(folder_path, file))
    
    if not images:
        print(f"No images found in {folder_path}")
        return
    
    print(f"Found {len(images)} images. Processing...")
    
    # Process each image
    results = []
    for img_path in images:
        print(f"\nProcessing {os.path.basename(img_path)}:")
        is_pothole, confidence = predict_image_file(img_path, model)
        results.append((img_path, is_pothole, confidence))
    
    # Print summary
    print("\nSummary:")
    pothole_count = sum(1 for _, is_pothole, _ in results if is_pothole)
    print(f"Total images: {len(results)}")
    print(f"Images with potholes: {pothole_count} ({pothole_count/len(results):.2%})")

def main():
    """Main function"""
    model = load_pothole_model(MODEL_PATH)
    
    if model is None:
        print("Error: Could not load model. Exiting...")
        return
    
    # Create results directory if it doesn't exist
    os.makedirs('results', exist_ok=True)
    
    while True:
        print("\nPothole Detection Menu:")
        print("1. Test on a single image")
        print("2. Test on a folder of images")
        print("3. Test on a video file")
        print("4. Exit")
        
        choice = input("Enter your choice (1-4): ")
        
        if choice == '1':
            image_path = input("Enter path to the image file: ")
            if os.path.exists(image_path):
                predict_image_file(image_path, model)
            else:
                print(f"Error: File {image_path} not found")
                
        elif choice == '2':
            folder_path = input("Enter path to the folder containing images: ")
            if os.path.isdir(folder_path):
                process_image_folder(folder_path, model)
            else:
                print(f"Error: Folder {folder_path} not found")
                
        elif choice == '3':
            video_path = input("Enter path to the video file: ")
            if os.path.exists(video_path):
                output_path = os.path.join('results', f"output_{os.path.basename(video_path)}")
                
                # Ask for processing parameters
                try:
                    confidence = float(input("Enter confidence threshold (0.1-0.9, default 0.6): ") or "0.6")
                    confidence = max(0.1, min(0.9, confidence))
                    
                    process_frames = int(input("Process every N frames (1-10, default 2): ") or "2")
                    process_frames = max(1, min(10, process_frames))
                except ValueError:
                    print("Invalid input, using default values.")
                    confidence = 0.6
                    process_frames = 2
                
                process_video(video_path, model, output_path=output_path, 
                             confidence_threshold=confidence, 
                             process_every_n_frames=process_frames)
            else:
                print(f"Error: File {video_path} not found")
                
        elif choice == '4':
            print("Exiting...")
            break
            
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
